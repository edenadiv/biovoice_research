# Evaluation Dashboard

This notebook gathers the core alpha metrics and the most important decision-level figures from the latest fusion run.


## Recommended Reading Order
1. Metrics JSON
2. Confusion matrices
3. Score distributions
4. Threshold sweep and calibration plots
5. Segment-level plots


```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import Image, display, Markdown

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
RUNS = sorted((ROOT / 'outputs' / 'runs').glob('*fusion_plus_interpretable_features'))
LATEST_RUN = RUNS[-1] if RUNS else None
print('Latest fusion run:', LATEST_RUN)
```

```python
metrics = json.loads((LATEST_RUN / 'metrics.json').read_text()) if LATEST_RUN else {}
metric_summary = pd.read_csv(LATEST_RUN / 'tables' / 'metric_summary.csv') if LATEST_RUN else pd.DataFrame()
metrics, metric_summary.head()
```

```python
if LATEST_RUN:
    for name in [
        'confusion_matrix.png',
        'normalized_confusion_matrix.png',
        'sv_roc.png',
        'spoof_roc.png',
        'target_vs_non_target_scores.png',
        'spoof_vs_bonafide_scores.png',
        'reliability_diagram.png',
        'threshold_sweep_heatmap.png',
        'score_scatter.png',
    ]:
        path = LATEST_RUN / 'plots' / name
        if path.exists():
            display(Image(filename=str(path)))
```

## Interpretation Guidance
The confusion matrices answer whether the final decision policy is behaving sensibly. The threshold heatmap and reliability diagram answer whether the current operating policy is brittle. The score scatter and histograms answer whether fusion is visually justified.
