# Baseline Model Review

This notebook reviews the compact speaker-verification and spoof-detection baselines used for alpha approval.


## Why These Baselines Matter
The goal is not to claim strong absolute performance. The goal is to demonstrate a clean, inspectable baseline that can later be replaced by stronger pretrained backbones without changing the surrounding research infrastructure.


```python
from pathlib import Path
import json
from IPython.display import Image, display, Markdown

ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
sv_runs = sorted((ROOT / 'outputs' / 'runs').glob('*sv_only'))
spoof_runs = sorted((ROOT / 'outputs' / 'runs').glob('*spoof_only'))
latest_sv = sv_runs[-1] if sv_runs else None
latest_spoof = spoof_runs[-1] if spoof_runs else None
print('Latest SV run:', latest_sv)
print('Latest spoof run:', latest_spoof)
```

```python
if latest_sv:
    display(Markdown('### Speaker Verification Baseline'))
    display(Image(filename=str(latest_sv / 'plots' / 'sv_loss_curves.png')))
if latest_spoof:
    display(Markdown('### Spoof Detection Baseline'))
    display(Image(filename=str(latest_spoof / 'plots' / 'spoof_loss_curves.png')))
```

Use these curves to verify that the training workflow is functioning and that train/validation behavior is stable enough for alpha experimentation. Do not read them as evidence of benchmark competitiveness.
